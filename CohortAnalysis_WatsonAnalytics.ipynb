```python
import pandas as pd
```

```python
import numpy as np
```

```python
ST = pd.read_excel("/Users/Niklas/Desktop/Master ESCP/Watson Analytics/Supermarket Transactions_Merged_FINAL.xlsx")
```

```python
ST.rename(columns={"Purchase Date" : "date"}, inplace = True)
```

```python
ST.head()
```

```python
ST["date"] = pd.to_datetime(ST["date"])
```

```python
ST["order_month"] = ST.date.dt.year.astype(str) + ST.date.dt.month.astype(str)
```

```python
ST.head()
```

```python
min_dates = ST.groupby("Customer ID").agg({"date":"min"}).reset_index()
min_dates["min_order_month"] = min_dates.date.dt.year.astype(str) + min_dates.date.dt.month.astype(str)
min_dates
```

```python
ST = ST.merge(min_dates, how="left", on="Customer ID", suffixes=["", "_y"])
ST
```

```python
ST_cohort = ST.groupby(["order_month", "min_order_month"]).agg({"Revenue":"sum"}).unstack()
ST_cohort
```
